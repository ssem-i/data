```python
import pandas as pd
import matplotlib.pyplot as plt

from google.colab import drive
drive.mount('/content/drive')


b = pd.read_csv('/content/drive/MyDrive/buylov.csv')
b
```

```python

bp = b['BGiftPrice']
plt.hist(bp)
plt.show()
```

```python
bp = b['BGiftPrice']
plt.boxplot(bp)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#b = pd.read_csv('buylov.csv')
bg = b[b['Role']=='Giver']
br = b[b['Role']=='Receiver']
print('<준  사람 선물가격>')
print('평균 :', np.mean(bg['BGiftPrice']))
print('중앙값:', np.median(bg['BGiftPrice']))
print('<받은 사람 선물가격>')
print('평균 :', np.mean(br['BGiftPrice']))
print('중앙값:', np.median(br['BGiftPrice']))
plt.boxplot([bg['BGiftPrice'],br['BGiftPrice']], labels=['Giver','Receiver'])
plt.show()

```

```python
#산점도


bg = b[b['Role']=='Giver']
br = b[b['Role']=='Receiver']
plt.figure(1)
plt.scatter(bg['BGiftPrice'], bg['Apprec-7'])
plt.figure(2)
plt.scatter(br['BGiftPrice'], br['Apprec-7'])
plt.show()

```
